```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\renuk\Downloads\Ecommerce_Sales_Dataset_600_Rows.xlsx")
```

```python
df.head()
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
df['Final_Sales'] = df['Sales'] - (df['Sales'] * df['Discount_%'] / 100)
```

```python
df['Profit'] = df['Final_Sales'] - df['Cost']
```

```python
df['Profit_Margin_%'] = (df['Profit'] / df['Final_Sales']) * 100
```

```python
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month_name()
```

```python
bins = [18,25,35,45,60]
labels = ['18-25','26-35','36-45','46-60']

df['Age_Group'] = pd.cut(df['Customer_Age'], bins=bins, labels=labels)
```

```python
df.to_csv("cleaned_ecommerce_data.csv", index=False)
```
